# EDA com visualizações
Crie visualizações para resopnder as perguntas do teste abaixo deste notebook.

```python
# Carregue o conjunto de dados
import pandas as pd

wine_df = pd.read_csv('winequality_edited.csv')
wine_df.head(2)
```

### Histogramas para diversas características

```python
# Histograma para Acidez fixa
wine_df["fixed acidity"].plot(kind="hist", title="Acidez");
```

```python
# Histograma para Total de dióxido de enxofre
wine_df['total sulfur dioxide'].plot(kind="hist", title="Dióxido de Enxofre");
```

```python
# Histograma para pH
wine_df['pH'].plot(kind="hist", title="pH ");
```

```python
# Histograma para Álcool
wine_df['alcohol'].plot(kind="hist", title="Álcool");
```

### Diagrama de dispersão da qualidade associada a diversas características

```python
# qualidade vs Acidez volátil
wine_df.plot(kind='scatter', x='volatile acidity', y='quality');
```

```python
# qualidade vs Açucar residual
wine_df.plot(kind='scatter', x='residual sugar', y='quality');
```

```python
# qualidade vs pH
wine_df.plot(kind='scatter', x='pH', y='quality');
```

```python
# qualidade vs Álcool
wine_df.plot(kind='scatter', x='alcohol', y='quality');
```
